```python
import pandas as pd

# Load the CSV
df = pd.read_csv("your_file.csv")

# Get rows with repeated SKUs
repeated_rows = df[df.duplicated('SKU', keep=False)]

# Convert to NumPy array
repeated_array = repeated_rows.to_numpy()

print(repeated_array)
```

```python
import pandas as pd

df = pd.read_csv("cleaned_sku.csv")

# Count missing (NaN) values in the 'SKU' column
missing_count = df['SKU'].isna().sum()

print(missing_count)
```

```python
import pandas as pd

# Load the CSV
df = pd.read_csv("your_file.csv")

# Drop rows where SKU is NaN
df_cleaned = df.dropna(subset=['SKU'])

# Export to a new CSV file
df_cleaned.to_csv("cleaned_sku.csv", index=False)

print("✅ Cleaned CSV saved as 'cleaned_sku.csv'")
```
